```python
import google.datalab.storage as storage
from pathlib import Path

bucket = storage.Bucket('leo_font')

for file in Path('/ai_font/data/r40202/ttfs/').rglob('*.ttf'):
    blob = bucket.object(file.name)
    blob.write_stream(file.read_text(), 'text/plain')
    print("File {} uploaded to {}.".format(file.name,bucket.name))
```

```python
from google.cloud import storage
from pathlib import Path
from tqdm import tqdm

source_file_fd = 'data/r40202/ttfs/'    # GCP에 업로드할 파일 절대경로
dest_file_fd = "r40202/ttfs/"

storage_client = storage.Client()
bucket = storage_client.bucket('leo_font')

pbar = tqdm(Path(source_file_fd).rglob('*.ttf'))
for file in pbar:
    blob = bucket.blob(dest_file_fd+file.name)
    blob.upload_from_filename(source_file_fd+file.name)
    pbar.set_postfix(filename=file.name)
```
