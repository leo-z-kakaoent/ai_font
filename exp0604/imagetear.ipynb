```python
import os
import numpy as np
from PIL import Image
```

```python
fd = "/home/jupyter/ai_font/data/test/raw/플레이브밤비"
files = sorted([f for f in os.listdir(fd) if "__" in f])
```

```python
for f in files:
    keepgoing = True
    while keepgoing:
        a = Image.open(f"{fd}/{f}")
        b = np.array(a)
        ij = input().split(" ")
        torn = Image.fromarray(b[:int(ij[0]),:int(ij[1])])
        display(a)
        display(torn)
        action = input()
        if action == "q":
            keepgoing = False
        elif action == "s":
            keepgoing = False
            torn.save(f"{fd}/{f.replace('.png','2.png')}")
```
