```python
import os
import pandas as pd
```

```python
path = "/home/jupyter/ai_font/data"

letter_mapper_a = pd.read_pickle(f"{path}/pickle/letter_mapper_a.pickle")
letter_mapper_b = pd.read_pickle(f"{path}/pickle/letter_mapper_b.pickle")
font_mapper = pd.read_pickle(f"{path}/pickle/font_mapper.pickle")
letter2font = pd.read_pickle(f"{path}/pickle/letter2font.pickle")
```

```python
sorted(os.listdir("/home/jupyter/ai_font/data/train"))
```

```python
fonts = []
letters_in = []
for f in os.listdir(f"{path}/train"):
    fonts.append(f)
    letters_in.append(sorted(list(set([p.replace(".png","").split("__")[-1] for p in os.listdir(f"{path}/train/{f}") if p.endswith(".png")]))))
```

```python
font_mapper = pd.Series(index=fonts, data=letters_in)
```

```python
font_mapper.to_pickle(f"{path}/pickle/font_mapper.pickle")
```

```python
from tqdm import tqdm
from collections import defaultdict
```

```python
letter2font = defaultdict(list)
```

```python
for font, letters in tqdm(zip(fonts, letters_in)):
    for letter in letters:
        letter2font[letter].append(font)
```

```python
pd.Series(letter2font)
```

```python
pd.Series(letter2font).to_pickle(f"{path}/pickle/letter2font.pickle")
```
