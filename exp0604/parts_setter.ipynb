```python
import os
import shutil
from tqdm import tqdm

path = "/home/jupyter/ai_font/data"
parts = "ㄱㄲㄴㄷㄸㄹㅁㅂㅃㅅㅆㅇㅈㅉㅊㅋㅌㅍㅎㅏㅐㅑㅒㅓㅔㅕㅖㅗㅘㅙㅚㅛㅜㅝㅞㅟㅠㅡㅢㅣㄳㄵㄶㄺㄻㄼㄽㄾㄿㅀㅄ"
```

```python
for font in tqdm(os.listdir(f"{path}/train")):
    os.makedirs(f"{path}/train_parts/{font}", exist_ok=True)
    for tag in ["closing","erode","dilate"]:
        for part in parts:
            from_path = f"{path}/train/{font}/{font}__{tag}__{part}.png"
            if os.path.exists(from_path):
                shutil.move(from_path, f"{path}/train_parts/{font}/{font}__{tag}__{part}.png")
```

```python
for font in tqdm(os.listdir(f"{path}/train")):
    for f in os.listdir(f"{path}/train/{font}"):
        if f.endswith(".png"):
            letter = f.replace(".png","").split("__")[-1]
            if letter in parts:
                os.remove(f"{path}/train/{font}/{f}")
```
